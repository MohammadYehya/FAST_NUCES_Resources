```python
#write a code to append an empty list using while loop.
n = 1
N = 10
l1 = []
while n < N:
  val = eval(input("Enter number: "))
  n+=1
  if val%10 == 0: continue
  l1.append(val)
print(l1)
#write a code to append an empty list using for loop.
l2 = []
for n in range(1,N):
  val = eval(input("Enter number: "))
  if val%10 == 0: break
  l2.append(val)
print(l2)
#Do use continue or break statement in above code.
```

```python
# write the code of to save a mathematical function and plot the function using matplotlib.pyplot.
# use function f(m)= sqrt(9.81*m/0.25)*tanh(4*sqrt(9.81*0.25/m))-36 don't forget to label the plot
import matplotlib.pyplot as plt
from numpy import arange,tanh,sqrt
m = arange(0.000001 , 20 , 1)
plt.plot(m,sqrt(9.81*m/0.25)*tanh(4*sqrt(9.81*0.25/m))-36,'o-')
plt.xlabel('m')
plt.ylabel('sqrt(9.81*m/0.25)*tanh(4*sqrt(9.81*0.25/m)')
plt.grid(True)
```
